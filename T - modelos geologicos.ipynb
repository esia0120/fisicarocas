# Modelos

```python
import numpy as np
from matplotlib import pyplot as plt
```

## Modelo de capas planas

```python
x = 90
z = 50
matrix = np.ones((z,x))
matrix[25:,:] = 2

fig, ax = plt.subplots()
im = plt.imshow(matrix, cmap='Wistia')
cbar = plt.colorbar(im, shrink=.6, ticks=np.unique(matrix))
cbar.set_label('Capas')

plt.xlim([0, 89])
plt.ylim([49, 0])

fig.set_dpi(300)
plt.xlabel('x')
plt.ylabel('z')
plt.title('Modelo de 2 Capas')
plt.show()
```

## Modelo de falla

```python
x = 90
z = 50
matrix  = np.zeros((z,x))

h1 = 0.4
h2 = 0.6
w1 = 0.25
for i in range(z):
    for j in range(x):
        if j/x < w1:
            h = h2
        elif ((1-2*w1)*(i+1))/((j+1)-w1*x) < z/x:
            h = h1
        else:
            h = h2

        if (i+1)/z < h:
            matrix[i][x-j-1] = 1
        else:
            matrix[i][x-j-1] = 2

fig, ax = plt.subplots()
im = plt.imshow(matrix, cmap='Wistia')
cbar = plt.colorbar(im, shrink=.6, ticks=np.unique(matrix))
cbar.set_label('Capas')

# Falla 
y = [0, 49]
x = [66, 23]
plt.plot(x, y, 'k')
ax.annotate("",
            xy=(50, 15), xycoords='data',
            xytext=(37, 30), textcoords='data',
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3"))

ax.annotate("",
            xy=(54, 18), xycoords='data',
            xytext=(42, 32), textcoords='data',
            arrowprops=dict(arrowstyle="<-", connectionstyle="arc3"))

plt.xlim([0, 89])
plt.ylim([49, 0])

fig.set_dpi(300)
plt.xlabel('x')
plt.ylabel('z')
plt.title('Falla')

plt.show()
```

## Modelo de graben

```python
x = 90
z = 50
matrix  = np.zeros((z,x))

h1 = 0.5
h2 = 0.6
w1 = 0.4

for i in range(z):
    for j in range(x):
        if (-w1*(i+1)/(j-x) >= z/x) or (w1*(i+1)/(j+1) >= z/x):
            h = h1
        else:
            h = h2
        if ((i+1)/z)<h:
            matrix[i][x-j-1] = 1
        else:
            matrix[i][x-j-1] = 2

fig, ax = plt.subplots()
im = plt.imshow(matrix, cmap='Wistia')
cbar = plt.colorbar(im, shrink=.6, ticks=np.unique(matrix))
cbar.set_label('Capas')

# Falla 1
y = [0, 49]
x = [0, 35]
plt.plot(x, y, 'k')
ax.annotate("",
            xy=(21, 33), xycoords='data',
            xytext=(12, 21), textcoords='data',
            arrowprops=dict(arrowstyle="<-", connectionstyle="arc3"))

ax.annotate("",
            xy=(27, 35), xycoords='data',
            xytext=(17, 21), textcoords='data',
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3"))

# Falla 2
y = [0, 49]
x = [89, 54]
plt.plot(x, y, 'k')
ax.annotate("",
            xy=(62, 35), xycoords='data',
            xytext=(72, 21), textcoords='data',
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3"))

ax.annotate("",
            xy=(66, 35), xycoords='data',
            xytext=(76, 21), textcoords='data',
            arrowprops=dict(arrowstyle="<-", connectionstyle="arc3"))

plt.xlim([0, 89])
plt.ylim([49, 0])

fig.set_dpi(300)
plt.xlabel('x')
plt.ylabel('z')
plt.title('Graben')

plt.show()
```
