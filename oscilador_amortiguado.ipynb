```python
import ipywidgets as widgets
import numpy as np
from matplotlib import pyplot as plt
plt.style.use(['science', 'notebook', 'grid'])
```

```python
#Valores iniciales
x0=0.08 #posicion
dx0=0.7 #velocidad
```

```python
def osc_amor(m=2,k=10,f=0.5):
        t=np.linspace(0,10,101)
        r1=(-f-((f**2)-4*m*k)**(1/2))/(2*m)
        r2=(-f+((f**2)-4*m*k)**(1/2))/(2*m)
        A=np.array([[1,1],[r1, r2]])
        B=np.array([[x0],[dx0]])
        sol=np.dot(np.linalg.inv(A),B)
        c1=sol[0]
        c2=sol[1]
        xt=c1*np.exp(r1*t) + c2*np.exp(r2*t)
        vx=r1*c1*np.exp(r1*t) + r2*c2*np.exp(r2*t)
        fig, ax=plt.subplots()
        plt.plot(t,xt,label='Posicionamiento')
        plt.plot(t,vx, label='Velocidad')
        plt.title('Oscilador Amortiguado')
        plt.xlabel('Tiempo [s]')
        plt.ylabel('Magnitud')
        legend = ax.legend(loc='upper right', shadow=True, fontsize='large')
        plt.xlim(0,10)
        plt.ylim(-0.4,0.4)
        plt.show()
```

```python
osc_amor(k=30)
```

```python
widgets.interact(osc_amor,m=(0.5,2,0.1),k=(10,30,1),f=(0.5,1.5,0.1));
```
